```python
def validate_recharge(amount, plan):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif plan not in valid_plans:
        return "Invalid plan selected."
    else:
        return "Recharge Successful!"

# Retry using while loop
while True:
    amount = float(input("Enter recharge amount: "))
    plan = int(input("Enter plan (199, 299, 399, 599): "))
    
    result = validate_recharge(amount, plan)
    print(result)
    
    if result == "Recharge Successful!":
        break
```

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Example dictionary
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12,
    "Mangoes": 30
}

check_inventory(inventory)
```

```python
def process_result(marks_list):
    total = 0
    
    for mark in marks_list:
        total += mark
        
    average = total / len(marks_list)
    
    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"

# Example usage
marks = [60, 55, 70, 45]
print(process_result(marks))
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if peak.lower() == "yes":
        fare += fare * 0.25
        
    return fare

# Retry option
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    print(f"Total Fare: ₹{total_fare:.2f}")
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```
